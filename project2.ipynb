## Part 2: Explore one Wikipedia page with the beautifulsoup package

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
#save and print the text content of a page with all tags removed

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
}

url = "https://en.wikipedia.org/wiki/LeBron_James"
response = requests.get(url, headers=headers) 

soup = BeautifulSoup(response.text, "html.parser")

text_content = soup.get_text()

with open("lebron_james.txt", "w", encoding="utf-8") as file:
    file.write(text_content)

print(text_content)  


#I attempted the code with the demo code given on Professor's github. However Wikipedia blocked me and sent me straight to the homepage so I could only print the homepage's text. I used chatgpt's help to mimic a browser so that I could still access the correct webpage. The output is as follows.

```

```python
#download an image with beautifulsoup and save it in this repository

infobox = soup.find("table", {"class": "infobox"})  
if infobox:
    img_tag = infobox.find("img") 

    if img_tag and 'src' in img_tag.attrs:
        img_url = img_tag['src']

        if img_url.startswith("//"):
            img_url = "https:" + img_url
        elif img_url.startswith("/"):
            img_url = "https://en.wikipedia.org" + img_url

        img_response = requests.get(img_url, headers=headers)

        with open("lebron_james.jpg", "wb") as file:
            file.write(img_response.content)

        print(f"Image saved as lebron_james.jpg from {img_url}")
    else:
        print("No image found in the infobox.")
else:
    print("Infobox not found.")
```

```python
#find all the links in a page with beautifulsoup
#print the first 100 characters of ten of these links

import requests
from bs4 import BeautifulSoup

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
}

url = "https://en.wikipedia.org/wiki/LeBron_James"
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")

link_url = [a['href'] for a in soup.find_all('a', href=True) if not a['href'].startswith('#')]

count = 0
for i in link_url[:10]:
  
    full_url = f"https://en.wikipedia.org{i}" if i.startswith('/') else i
    
    source = requests.get(full_url, headers=headers).text
    soup = BeautifulSoup(source, "html.parser")
    
    page_text = soup.get_text()
    print(f"Link {count + 1}: {full_url}")
    print(f"First 100 characters: {page_text[:100]}")
    print("-" * 80)
    
    count += 1
    if count == 10:
        break
```

```python
#using the links in the "link" column, download the first 1000 characters of each script
#use requests and bs4, remember to remove all html tags
import pandas as pd
import requests
from bs4 import BeautifulSoup

df = pd.read_csv("pudding_data.csv")


def get_script_text(url):
    try:
        response = requests.get(url, timeout=10)
        response.raise_for_status()
        
        if "text/html" not in response.headers.get("Content-Type", ""):
            return "Non-HTML content"

        response.encoding = response.apparent_encoding 
        soup = BeautifulSoup(response.text, "html.parser") 
        
        text = soup.get_text() 
        text = text.replace("\n", " ").replace("\t", " ")  
        
        return text[:1000]
    except requests.RequestException:
        return "Error fetching script"


df["script_preview"] = df["link"].apply(get_script_text)

df.to_csv("pudding_data_with_scripts.csv", index=False)




```

```python
#download the movie posters for 10 of these movies and save them to this repository

base_url = "https://api.themoviedb.org/3/movie/"
image_base_url = "https://image.tmdb.org/t/p/original"
headers = {
    "accept": "application/json",
    "Authorization": "Bearer eyJhbGciOiJIUzI1NiJ9.eyJhdWQiOiJlNTEwNjkwZTA4MzczMTNhZjUzMWY5NTdjYWNmYzcwZCIsIm5iZiI6MTc0MDUwODA3Ny4wOTksInN1YiI6IjY3YmUwYmFkMzljYjc4MjBkMGVmOTdmNSIsInNjb3BlcyI6WyJhcGlfcmVhZCJdLCJ2ZXJzaW9uIjoxfQ.EgWSOmyOFk64D0PmwnMr5CunEubUcgtzt2DAMIsG-gg"
}

movies = {
    "Mickey 17": 696506,
    "The Substance": 933260,
    "Conclave": 974576,
    "The Wild Robot": 1184918,
    "The Brutalist": 549509,
    "Wicked": 402431,
    "Panda Plan": 1160956,
    "Flight Risk": 1126166,
    "Anora": 1065213,
    "The Gorge": 950396
}

def download_poster(movie_name, movie_id):
    response = requests.get(f"{base_url}{movie_id}", headers=headers)
    
    if response.status_code == 200:
        data = response.json()
        poster_path = data.get("poster_path")
        
        if poster_path:
            poster_url = f"{image_base_url}{poster_path}"
            img_data = requests.get(poster_url).content
            
            filename = f"{movie_name.replace(' ', '_')}.jpg"
            with open(filename, "wb") as file:
                file.write(img_data)
            print(f"✅ Saved: {filename}")

for name, movie_id in movies.items():
    download_poster(name, movie_id)
```
